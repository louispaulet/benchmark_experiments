## create dataset  

Sums starting at 1+1.  
Results from 2 to 100.  
99 rows.  

```python
import pandas as pd

addition_list = []
ones_list = []
for i in range(100):
  ones_list.append('1')
  addition_list.append({
      'sum' : '+'.join(ones_list),
      'result' : str(i+1)
      })

additions_df = pd.DataFrame(addition_list)

# we remove the first line that doesn't contain a sum (single one)
additions_df = additions_df[additions_df['result'] != '1']
additions_df.reset_index(drop=True, inplace=True)

additions_df.head()
```

```python
from datasets import Dataset

additions_ds = Dataset.from_pandas(additions_df, preserve_index=False)
additions_ds
```
